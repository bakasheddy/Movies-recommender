```python
import ast
def extract_known_for_info(known_for_list):
    result = []
    for item in known_for_list:
        result.append({
            'backdrop_path': item.get('backdrop_path'),
            'id': item.get('id'),
            'title': item.get('title'),
            'original_title': item.get('original_title'),
            'overview': item.get('overview'),
            'poster_path': item.get('poster_path'),
            'media_type': item.get('media_type'),
            'adult': item.get('adult'),
            'original_language': item.get('original_language'),
            'genre_ids': item.get('genre_ids'),
            'popularity': item.get('popularity'),
            'release_date': item.get('release_date'),
            'video': item.get('video'),
            'vote_average': item.get('vote_average'),
            'vote_count': item.get('vote_count')
        })
    return result

# Parse 'Known_for' column and expand it
df1['known_for'] = df1['known_for'].apply(ast.literal_eval)  # Convert string representation of list to actual list
df1 = df1['known_for'].apply(pd.Series).stack().reset_index(level=1, drop=True)
df1 = pd.DataFrame(df1.tolist())
```
